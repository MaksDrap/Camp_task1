```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randint(1, 101, size=20)
min_index = np.argmin(a)

plt.plot(a) 

left = min_index
right = min_index

while left > 0 and a[left-1] > a[left]:
    left -= 1

while right < len(a) - 1 and a[right+1] > a[right]:
    right += 1

plt.plot(range(left, right+1), a[left:right+1], 'r-') # 'r-' означає червону лінію

plt.show()
```
